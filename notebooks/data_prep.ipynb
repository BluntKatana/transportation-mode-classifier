# Combine data from single experiment into single dataframe
We have the following data from a single experiment: Accelerometer.csv, Gyroscope.csv, Light.csv, Linear Acceleration.csv, Location.csv, Magnetometer.csv, Pressure.csv, and Temperature.csv.


```python
# Define the folder path and load the data
FOLDER = "../data/bike/Fiets 2024-06-05 10-03-06/"
TRANSPORT_MODE = "bike"
EXPIREMENT_ID = np.random.randint(0, 1000000)
START_DATE = pd.read_csv(FOLDER + "meta/time.csv", delimiter=";").iloc[0, 3]
SYSTEM_TIME = pd.read_csv(FOLDER + "meta/time.csv", delimiter=";").iloc[0, 2]

data_files = {
    "accelerometer": "Accelerometer.csv",
    "gyroscope": "Gyroscope.csv",
    "light": "Light.csv",
    "linear_acceleration": "Linear Acceleration.csv",
    "location": "Location.csv",
    "magnetometer": "Magnetometer.csv",
    "pressure": "Pressure.csv",
    "proximity": "Proximity.csv",
    "temperature": "Temperature.csv",
}

dataframes = {}
for name, file in data_files.items():
    # check if file exists
    try:
        dataframes[name] = pd.read_csv(FOLDER + file, delimiter=";")

        # Convert "Time (s)" column from scientific to float notation
        dataframes[name]["Time (s)"] = dataframes[name]['Time (s)'].str.replace(',', '.').replace("E", "e")
        dataframes[name]["Time (s)"].astype(float)
    except FileNotFoundError:
        print(f"File {file} not found. Skipping...")
        continue

print(START_DATE)
print(SYSTEM_TIME)
```

```python
# Merge dataframes based on the "Time (s)" column
merged_df = None
for name, df in dataframes.items():
    if merged_df is None:
        merged_df = df
    else:
        # Convert "Time (s)" column from scientific to float notation
        merged_df = pd.concat([merged_df, df])
        # merged_df = merged_df.merge(df, how='outer', on='Time (s)')

# Display statistics of time (s) column
merged_df["Time (s)"] = merged_df["Time (s)"].str.replace(',', '.').replace("E", "e")

# Convert "Time (s)" column to float
merged_df["Time (s)"] = merged_df["Time (s)"].astype(float)

# Display the merged dataframe
merged_df.head()
```

```python
# Create a new column "timestamp" based on the "Time (s)" column and the start date
merged_df["timestamp"] = pd.to_timedelta(merged_df["Time (s)"], unit='s') + pd.to_datetime(START_DATE)
merged_df["transportation_mode"] = TRANSPORT_MODE
merged_df["start_date"] = START_DATE
merged_df["expirement_id"] = EXPIREMENT_ID
```
